```python
mos_hex = read("surfaces/MoS2_hex_1x1_1layer.xyz")
print(f"Nat = {len(mos_hex)}")
print(f"Cell =\n{mos_hex.get_cell()}")
```

```python
mos_ortho = make_supercell(mos_hex, [[1, 1, 0], [-1, 1, 0], [0, 0, 1]])
print(f"Nat = {len(mos_ortho)}")
print(f"Cell =\n{mos_ortho.get_cell()}")
```

```python
write("surfaces/MoS2_ortho_1x1_1layer.xyz", mos_ortho)
```

```python
mos_ortho_6by6 = mos_ortho*(6, 4, 1)
print(f"Nat = {len(mos_ortho_6by6)}")
print(f"Cell =\n{mos_ortho_6by6.get_cell()}")
```

```python
write("surfaces/MoS2_ortho_6x4_1layer.xyz", mos_ortho_6by6)
```
